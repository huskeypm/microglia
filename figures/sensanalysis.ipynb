```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt
```

```python
cs = 10**np.linspace(-6,0,10)
```

I'd like to provide some predictions of how partial (ant)agonism of key players identified from sensitivity analysis influence TNFa. Here's one idea of how to represent this 

```python
f, (l,m,r) = plt.subplots(1,3)

### P2X
l.plot(cs,1/(1 + 1e-3/cs),label="ctl")
l.plot(cs,1/(1 + 1e-4/cs),label="overexpress")
l.plot(cs,1/(1 + 1e-5/cs),label="inhib")

l.set_ylabel("TNF?")
l.set_xlabel("[ATP]")
l.set_xscale('log')
l.set_title("P2X")
l.legend(loc=0)
l.spines['right'].set_visible(False)
l.spines['top'].set_visible(False)


### NFAT
m.set_title("NFAT")
m.plot(cs,1/(1 + 1e-3/cs),label="ctl")
m.set_xscale('log')

def fancypants(m):
    m.spines['left'].set_visible(False)
    m.spines['right'].set_visible(False)
    m.spines['top'].set_visible(False)
    m.tick_params(
        axis='y',          # changes apply to the x-axis
        which='both',      # both major and minor ticks are affected
        left='off',      # ticks along the bottom edge are off
        top='off',         # ticks along the top edge are off
        labelleft='off') # labels along the bottom edge are off
fancypants(m)
    
### p38
r.set_title("p38")
r.plot(cs,1/(1 + 1e-3/cs),label="ctl")
r.set_xscale('log')

fancypants(r)


plt.tight_layout()
#plt.xscale('log')
```
