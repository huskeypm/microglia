```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
x = np.array([1,10,100,1000])
y = np.array([50,150,450,900])
X, Y = np.meshgrid(x, y)
Z = np.array([[ 0.14326171,  0.15356277,  0.16245797,  0.31226608],
              [ 0.1394732,   0.15094029,  0.1591781,   0.26757278],
              [ 0.13941233,  0.14856883,  0.15376791,  0.26167386],
              [ 0.1393419,   0.14609531,  0.14860826,  0.2616689 ]])

plt.figure(figsize=(7,8))
plt.tick_params(labelsize=12)
im = plt.imshow(Z, interpolation='bilinear', origin='lower',
                cmap=cm.gray, extent=(1, 1000, 50, 900),aspect='auto')
levels = np.arange(0.11, 0.26, 0.02)
CS = plt.contour(Z, levels,
                 origin='lower',
                 colors=('y','palegreen','paleturquoise','mediumspringgreen','lawngreen','gold'),
                 linewidths=2,
                 extent=(1, 1000, 50, 900))

CBI = plt.colorbar(im, orientation='horizontal', shrink=0.8)
CBI.set_label('$[Ca^{2+}]_{i}$ conc. (uM)', fontsize=12)
plt.clabel(CS, inline=1, fontsize=12)
plt.xlabel("ATP conc. (uM)",fontsize=15)
plt.ylabel("Duration (sec)",fontsize=15)
plt.tight_layout()
plt.savefig("/home/AD/bch265/Dropbox/Postdoctoral2017-/Modeling-Codes/repo/bens-paper/microglia04092018/contCaduravsATP.png")
```
