```python
stateLabel = "Cai"
subData1 = ao.GetData(data1,stateLabel)
subData2 = ao.GetData(data2,stateLabel)
subData3 = ao.GetData(data3,stateLabel)
subData4 = ao.GetData(data4,stateLabel)
subData5 = ao.GetData(data5,stateLabel)

plt.figure(figsize=(7,10))
ax = plt.subplot(5,1,1)
ax.tick_params(labelsize=12)
ax.plot(subData1.t*1000-1050,subData1.valsIdx,'b-',label="ST50/RS50")
ax.legend(fontsize=12)
ax.set_ylabel("[Ca]i (uM)",fontsize=15)
#plt.ylim(0.05,0.25)
plt.xlim(2000,4000)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
ax.set_xticklabels(empty_string_labels)
plt.tight_layout()

ax = plt.subplot(5,1,2)
ax.tick_params(labelsize=12)
ax.plot(subData2.t*1000-1050,subData2.valsIdx,'r-',label="ST50/RS100")
ax.legend(fontsize=12)
ax.set_ylabel("[Ca]i (uM)",fontsize=15)
#plt.ylim(0.05,0.25)
plt.xlim(2000,4000)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
ax.set_xticklabels(empty_string_labels)
plt.tight_layout()

ax = plt.subplot(5,1,3)
ax.tick_params(labelsize=12)
ax.plot(subData3.t*1000-1050,subData3.valsIdx,'g-',label="ST50/RS200")
ax.legend(fontsize=12)
ax.set_ylabel("[Ca]i (uM)",fontsize=15)
#plt.ylim(0.05,0.25)
plt.xlim(2000,4000)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
ax.set_xticklabels(empty_string_labels)
plt.tight_layout()


ax = plt.subplot(5,1,4)
ax.tick_params(labelsize=12)
ax.plot(subData4.t*1000-1050,subData4.valsIdx,'m-',label="ST50/RS300")
ax.legend(fontsize=12)
ax.set_ylabel("[Ca]i (uM)",fontsize=15)
#plt.ylim(0.05,0.25)
plt.xlim(2000,4000)
labels = [item.get_text() for item in ax.get_xticklabels()]
empty_string_labels = ['']*len(labels)
ax.set_xticklabels(empty_string_labels)
plt.tight_layout()


ax = plt.subplot(5,1,5)
ax.tick_params(labelsize=12)
ax.plot(subData5.t*1000-3050,subData5.valsIdx,'k-',label="ST50/RS900")
ax.legend(fontsize=12)
ax.set_ylabel("[Ca]i (uM)",fontsize=15)
#plt.ylim(0.05,0.25)
plt.xlim(0,2000)
ax.set_xlabel("time (sec)",fontsize=15)
plt.tight_layout()

plt.savefig("/home/AD/bch265/Dropbox/Postdoctoral2017-/Modeling-Codes/repo/bens-paper/microglia04092018/cavsfreq.png")
```

```python
ca1 = max(subData1.valsIdx[2000:4000])
ca2 = max(subData2.valsIdx[2000:4000])
ca3 = max(subData3.valsIdx[2000:4000])
ca4 = max(subData4.valsIdx[2000:4000])
ca5 = max(subData5.valsIdx[2000:4000])

cas = np.array([ca1,ca2,ca3,ca4,ca5])
camax = max(cas)

car = cas/camax

rest = np.array([50,100,200,300,900])

plt.figure(figsize=(7,7))
plt.tick_params(labelsize=12)
plt.plot(rest,car,'k--s',label="ST50 - 1 mM ATP")
plt.xlabel("Resting term (sec)",fontsize=15)
plt.ylabel("Relative [Ca] (fraction of max)")

plt.ylim(0.4,1.02)
plt.xlim(0,920)

plt.xticks(np.arange(0, 910, 100))
plt.tight_layout()
plt.savefig("/home/AD/bch265/Dropbox/Postdoctoral2017-/Modeling-Codes/repo/bens-paper/microglia04092018/cavsfreqrelplot.png")
```
