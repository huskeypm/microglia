```python
CaData = np.array([max(CaData1.valsIdx),max(CaData2.valsIdx),max(CaData3.valsIdx),max(CaData4.valsIdx),
                   max(CaData5.valsIdx),max(CaData6.valsIdx),max(CaData7.valsIdx),max(CaData8.valsIdx),max(CaData9.valsIdx)])
CaERData = np.array([max(CaERData1.valsIdx),max(CaERData2.valsIdx),max(CaERData3.valsIdx),max(CaERData4.valsIdx),
                     max(CaERData5.valsIdx),max(CaERData6.valsIdx),max(CaERData7.valsIdx),max(CaERData8.valsIdx),max(CaERData9.valsIdx)])
NFATData = np.array([max(NFATData1.valsIdx),max(NFATData2.valsIdx),max(NFATData3.valsIdx),max(NFATData4.valsIdx),
                     max(NFATData5.valsIdx),max(NFATData6.valsIdx),max(NFATData7.valsIdx),max(NFATData8.valsIdx),max(NFATData9.valsIdx)])
pp38Data = np.array([max(pp38Data1.valsIdx),max(pp38Data2.valsIdx),max(pp38Data3.valsIdx),max(pp38Data4.valsIdx),
                     max(pp38Data5.valsIdx),max(pp38Data6.valsIdx),max(pp38Data7.valsIdx),max(pp38Data8.valsIdx),max(pp38Data9.valsIdx)])
CaMCNData = np.array([max(CaMCNData1.valsIdx),max(CaMCNData2.valsIdx),max(CaMCNData3.valsIdx),max(CaMCNData4.valsIdx),
                      max(CaMCNData5.valsIdx),max(CaMCNData6.valsIdx),max(CaMCNData7.valsIdx),max(CaMCNData8.valsIdx),max(CaMCNData9.valsIdx)])
TNFaData = np.array([max(TNFaData1.valsIdx),max(TNFaData2.valsIdx),max(TNFaData3.valsIdx),max(TNFaData4.valsIdx),
                     max(TNFaData5.valsIdx),max(TNFaData6.valsIdx),max(TNFaData7.valsIdx),max(TNFaData8.valsIdx),max(TNFaData9.valsIdx)])

Car = CaData/max(CaData)
CaERr = CaERData/max(CaERData)
NFATr = NFATData/max(NFATData)
pp38r = pp38Data/max(pp38Data)
CaMCNr = CaMCNData/max(CaMCNData)
TNFar = TNFaData/max(TNFaData)
```

```python
xaxis = [100, 95, 70, 60, 50, 40, 30, 5, 0]
plt.figure(figsize=(7,7))
plt.tick_params(labelsize=12)
plt.semilogx(xaxis,Car,'r-',linewidth=2,label="Ca")
plt.semilogx(xaxis,CaERr,'b-',linewidth=2,label="CaER")
plt.semilogx(xaxis,NFATr,'g-',linewidth=2,label="NFAT")
plt.semilogx(xaxis,pp38r,'m-',linewidth=2,label="p-p38")
plt.semilogx(xaxis,CaMCNr,'c-',linewidth=2,label="CaM-CN")
plt.semilogx(xaxis,TNFar,'k-',linewidth=2,label="TNFa released")
plt.xlabel("Relative Ca Flux Rate (%)",fontsize=15)
plt.ylabel("% of max",fontsize=15)
plt.legend(loc=0,fontsize=12)
plt.grid(True)
plt.tight_layout()
```
